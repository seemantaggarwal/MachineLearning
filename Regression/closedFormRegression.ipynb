```python
X,Y= make_regression(n_samples=400, n_features=1, n_informative=1, noise = 1.8, random_state = 11)
```

```python
Y=Y.reshape((-1,1))
```

```python
X = (X-X.mean())/X.std()
```

```python
plt.plot(X,Y)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.title("normalised data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
ones.shape
```

```python
X_=np.hstack((X,ones))
```

```python
def predict(X,theta):
    return np.dot(X,theta)
def getclosedform(X,Y):
    Y = np.mat(Y)
    first = np.dot(X.T, X)
    second = np.dot(X.T,Y)
    theta = np.linalg.pinv(first) * second
    return theta
X.shape
```

```python
theta = getclosedform(X_,Y)
theta
```

```python
plt.scatter(X,Y)
plt.plot(X, predict(X_, theta), color="red", label="prediction")
plt.show()
```

```python
loss = np.dot(((X_*theta) - Y).T, (X_*theta) - Y)
loss
```
