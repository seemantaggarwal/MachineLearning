```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =  np.arange(10)
y = x**2
y2= 2*x +3
```

```python
plt.plot(x,y, color='red', label='apple', marker='o')
plt.plot(x,y2,color='green', label='kiwi', linestyle='dashed')
plt.xlabel('time')
plt.ylabel('price')
plt.title('fruit over time')
plt.legend()
plt.show()
```

```python
price=np.array([1,2,34,5,6,4,35,5])
plt.plot(price)
```

```python
plt.plot(x,y,color='red', label='apple', marker='^')
plt.plot(x,y2, color='green', label='kiwi', marker='*')
plt.xlabel('time')
plt.ylabel('price')
```

```python
x_coor=np.array([1,2,3])*2
sx = x_coor+0.5
plt.bar(x_coor , [10,20,30],width = 0.5, label = 'last year', tick_label=["gold", "platinum", "silver"])
plt.bar(sx , [10,15,10],width=0.5, label ='current year')
plt.xlabel('metal')
plt.ylabel('price')
plt.ylim(0,40)
plt.legend()
plt.show()
```

```python
subs = ["maths", "english", "DS", "dicrete", "modulation"]
weights =[2,1,3,2,2]
plt.style.use("seaborn")
plt.pie(weights,labels=subs,  autopct='%1.1f%%', shadow='true')
plt.show()
```
